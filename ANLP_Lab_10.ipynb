Behind the scenes

```python
def translate_to_arabic(text):
    translator = Translator()
    try:
        translation = translator.translate(text, dest='ar')
        return translation.text
    except Exception as e:
        print(f"An error occurred during translation: {e}")
        return None
```

```python
def get_headlines(api_key):
    url = f"https://newsapi.org/v2/top-headlines?country=us&apiKey={api_key}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        headlines = [article['title'] for article in data['articles']]
        return headlines
    else:
        print("Error:", response.status_code)
        return []
```

```python
def get_flight_details(api_key, flight_number):
    url = f"https://api.aviationstack.com/v1/flights"
    params = {
        'access_key': api_key,
        'flight_iata': flight_number
    }

    # Making the API request
    response = requests.get(url, params=params)

    # Check if the request was successful
    if response.status_code == 200:
        data = response.json()
        if data['data']:
            flight_info = data['data'][0]
            details = {
                'airline': flight_info['airline']['name'],
                'departure_airport': flight_info['departure']['airport'],
                'arrival_airport': flight_info['arrival']['airport'],
                'departure_time': flight_info['departure']['estimated'],
                'arrival_time': flight_info['arrival']['estimated'],
                'status': flight_info['flight_status']
            }
            return details
        else:
            print("No flight information found.")
            return {}
    else:
        print(f"Error: {response.status_code}")
        return {}
```

```python
from IPython.display import Audio, display

def read_headlines(headlines):
    for i, headline in enumerate(headlines):
        print(headline)  # Display each headline
        tts = gTTS(text=headline, lang='en')
        filename = f"headline_{i}.mp3"
        tts.save(filename)

        # Use IPython.display.Audio to play the audio
        display(Audio(filename, autoplay=True))

        # Wait for a few seconds to allow the audio to finish playing
        time.sleep(5)

        # Optionally, remove the file after playing
        os.remove(filename)
```

## **Interface**

```python
def display_translation(original_text, translated_text):
    print("\n" + "=" * 40)
    print("Translation Summary")
    print("=" * 40)
    print(f"Original Text: {original_text}")
    print(f"Arabic Translation: {translated_text if translated_text else 'Translation Failed'}")
    print("=" * 40)

def main():
    while True:
        # User Input
        english_text = input("Enter the English text to translate (or type 'exit' to quit): ")

        if english_text.lower() == 'exit':
            print("Exiting the translation program. Goodbye!")
            break

        translated_text = translate_to_arabic(english_text)

        # Display Output
        display_translation(english_text, translated_text)

# Run the main function
main()
```

```python
# Get Headlines
def display_headlines(headlines):
    table = PrettyTable()
    table.field_names = ["Index", "Headline"]

    for index, headline in enumerate(headlines, start=1):
        table.add_row([index, headline])

    print(table)

# Example usage
headlines = get_headlines(api_key_news)
print("Headlines:")
display_headlines(headlines)
```

```python
# Get Flight Details
def display_flight_details(flight_details):
    table = PrettyTable()
    table.field_names = ["Attribute", "Details"]

    for key, value in flight_details.items():
        table.add_row([key.capitalize(), value])

    print(table)

# Example usage
flight_details = get_flight_details(api_key_flight, 'DL106')
print("Flight Details:")
display_flight_details(flight_details)
```
